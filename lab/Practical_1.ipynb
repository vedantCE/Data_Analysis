In this article we will study Pandas Series which is a useful one-dimensional data structure in Python.

```python
import pandas as pd

data = [1, 2, 3, 4]
 
ser = pd.Series(data)
print(ser)
```

```python
import pandas as pd
import numpy as np

data = np.array(['g','e','e','k','s','f', 'o','r','g','e','e','k','s'])
ser = pd.Series(data)

print(ser[:5])
```

```python
# Syntex:
# pandas.Series(data, index, dtype, copy)

# How to use:copy
# deep=True: Creates a Deep Copy. It copies both the indices and the actual data. Modifications to new_s will not affect s.
# deep=False: Creates a Shallow Copy. It only copies references to the data.

# Where to use:
# Experimental Work: When you want to try different filters or transformations without ruining your main dataset.

# Avoiding Warnings: Prevents the SettingWithCopyWarning when you extract a column from a DataFrame and want to modify it independently. 

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temps = np.array([30, 32, 31, 29, 28])
s = pd.Series(temps, index=days,dtype='float64',copy=True)
print(s)

# Copy Parameter કેવી રીતે વાપરવું:

# deep=True: Deep Copy બનાવે છે
# આ indices અને actual data બંને ની copy બનાવે છે
# નવા series (new_s) માં changes કરો તો original series (s) પર કોઈ અસર નહીં થાય
# બંને completely અલગ બની જાય છે - જેમ કે બે અલગ notebooks


# deep=False: Shallow Copy બનાવે છે

# આ ફક્ત data નો reference copy કરે છે, actual data નહીં
# એક જ data ને બે નામો આપે છે જેવું
# નવા series માં changes કરો તો original પણ change થઈ શકે છે

# કયારે વાપરવું (Where to use):
# 1. Experimental Work માટે:
# જ્યારે તમે different filters અથવા transformations try કરવા માંગો છો અને તમારું main dataset ખરાબ કરવું નથી. જેમ કે:
#Original data safe રાખવો હોય
# નવા experiments કરવા હોય પણ backup જોઈએ
# Multiple versions ની જરૂર હોય

# 2. Warnings ટાળવા માટે:
# SettingWithCopyWarning prevent કરે છે જ્યારે તમે DataFrame માંથી column extract કરો છો અને એને independently modify કરવું હોય. આ warning આવે છે જ્યારે pandas ને ખબર નથી કે તમે original data change કરો છો કે copy
```

What is a DataFrame?

A DataFrame is a two-dimensional labeled data structure (rows × columns).

Stores multiple columns

Each column is a Series

Has row index and column names

Similar to an Excel sheet / database table

```python
import pandas as pd

data = [1, 2, 3, 4]
 
ser = pd.DataFrame(data)
print(ser)
```
