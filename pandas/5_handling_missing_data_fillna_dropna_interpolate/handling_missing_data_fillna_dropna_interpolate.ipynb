## <font color="blue">fillna</font>

<font color="purple">**Fill all NaN with one specific value**</font>

```python
new_df = df.fillna(0)
new_df
```

<font color="purple">**Fill na using column names and dict**</font>

```python

#df.fillna(0)
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 1,
        'event': 'No Event'
    })
new_df
```

<font color="purple">**Use method to determine how to fill na values**</font>

```python
new_df = df.fillna(method="ffill") #limit, axis
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

<font color="purple">**Use of axis**</font>

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

<font color="purple">**limit parameter**</font>

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

### <font color="blue">interpolate</font>

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time") 
new_df
```

**Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)**

### <font color="blue">dropna</font>

```python
df
```

```python
new_df = df.dropna()
new_df
```

```python
df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1) #Only Keep Rows with Minimum Number of non-NaN Values
new_df
```

```python
#only keep rows with at least 2 non-NaN values
new_df = df.dropna(thresh=2) 
new_df
```

```python
#Only Keep Rows with Minimum % of non-NaN Values
df.dropna(thresh=0.3*len(df.columns))
```

##  Only Keep Columns with Minimum Number of non-NaN Values

```python
#only keep columns with at least 6 non-NaN values
df.dropna(thresh=7, axis=1)
```

```python
#only keep columns with at least 70% non-NaN values
df.dropna(thresh=0.7*len(df), axis=1)
```
