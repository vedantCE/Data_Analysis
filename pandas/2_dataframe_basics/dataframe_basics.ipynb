## <font color='blue'>Exercise Difference between loc(), iloc() and .ix() in Pandas DataFrame</font>

## loc: Label-based indexing. It is used to access rows and columns by their labels.
## iloc: Integer-location based indexing. It is used to access rows and columns by their integer positions.
## ix: ix supports mixed integer and label based access. (Deprecated)

```python
import pandas as pd

# Sample DataFrame
data = {
    'Occupation': ['Engineer', 'Doctor', 'Artist', 'Engineer', 'Doctor', 'Artist'],
    'Salary': [70000, 120000, 50000, 80000, 130000, 60000],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Education': ['Bachelors', 'Masters', 'Bachelors', 'Masters', 'Bachelors', 'Masters']
}

df = pd.DataFrame(data, index=['a', 'b', 'c', 'd', 'e', 'f'])
print("Original DataFrame:")
print(df)
```

```python
# Select a single row by label
row_b = df.loc['b']
print("\nRow 'b' using loc:")
print(row_b)

# Select multiple rows by label
rows_b_d = df.loc[['b', 'd']]
print("\nRows 'b' and 'd' using loc:")
print(rows_b_d)

# Select rows and columns by label
subset_loc = df.loc[['b', 'd'], ['Occupation', 'Salary']]
print("\nRows 'b' and 'd' with columns 'Occupation' and 'Salary' using loc:")
print(subset_loc)
```

```python
# Select a single row by position
row_1 = df.iloc[1]
print("\nRow 1 using iloc:")
print(row_1)

# Select multiple rows by position
rows_1_3 = df.iloc[[1, 3]]
print("\nRows 1 and 3 using iloc:")
print(rows_1_3)

# Select rows and columns by position
subset_iloc = df.iloc[[1, 3], [0, 1]]
print("\nRows 1 and 3 with columns 0 and 1 using iloc:")
print(subset_iloc)
```

```python
# Select a single row by position
row_1 = df.iloc[1]
print("\nRow 1 using iloc:")
print(row_1)

# Select multiple rows by position
rows_1_3 = df.iloc[[1, 3]]
print("\nRows 1 and 3 using iloc:")
print(rows_1_3)

# Select rows and columns by position
subset_iloc = df.iloc[[1, 3], [0, 1]]
print("\nRows 1 and 3 with columns 0 and 1 using iloc:")
print(subset_iloc)
```
