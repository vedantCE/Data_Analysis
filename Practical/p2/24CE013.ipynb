```python
import pandas as pd
import numpy as np

weather_data = {
    'Date': ['2026-01-01', '2026-01-02', '2026-01-03', '2026-01-04', '2026-01-05'],
    'City': ['Delhi', 'Delhi', 'Mumbai', 'Mumbai', 'Delhi'],
    'Temperature': [28, np.nan, 30, -5, np.nan],
    'Humidity': [65, 60, np.nan, 70, np.nan]
}

df = pd.DataFrame(weather_data)
df.to_csv("weather_missing.csv", index=False)

print(df)
```

```python
print(df.isnull())
print(df.isna())
print(df.notnull())
```

```python
print(df.isnull().sum())
```

```python
print(df[df['Temperature'].isnull()])
```

```python
print("Original Shape:", df.shape)
# Before dropna
```

```python
df_any = df.dropna()
print(df_any)
print("Shape after dropna():", df_any.shape)
# Remove rows with ANY missing values
```

```python
df_all = df.dropna(how='all')
# Remove rows where ALL values are missing
```

```python
# Remove rows where specific column has missing values
df_temp = df.dropna(subset=['Temperature'])
print(df_temp)
```

```python
# Filling Missing Data
df_filled = df.fillna(0)
print(df_filled)
```

```python
# Fill numeric columns with mean
temp_mean_np = np.mean(df['Temperature'].dropna())
hum_mean_pd = df['Humidity'].mean()

df['Temperature'] = df['Temperature'].fillna(temp_mean_np)
df['Humidity'] = df['Humidity'].fillna(hum_mean_pd)

print(df)
```

```python
# Forward Fill & Backward Fill
print(df.fillna(method='ffill'))
print(df.fillna(method='bfill'))
```

```python
# Create DataFrame with inconsistent labels
data = {
    'Name': ['Amit', 'Ravi', 'Neha', 'Karan'],
    'Gender': ['M', 'Male', 'male', np.nan],
    'Marks': [85, -10, 90, 300]
}

df_dirty = pd.DataFrame(data)
print("Before Cleaning:\n", df_dirty)
```

```python
df_dirty['Gender'] = df_dirty['Gender'].replace(
    ['M', 'Male', 'male'], 'Male'
)

df_dirty['Gender'] = df_dirty['Gender'].fillna('Unknown')
```

```python
df_dirty['Marks'] = df_dirty['Marks'].replace(
    [-10, 300], np.nan
)

df_dirty['Marks'] = df_dirty['Marks'].fillna(
    np.mean(df_dirty['Marks'])
)

print("After Cleaning:\n", df_dirty)
```

```python
grouped = df.groupby('City').agg(
    Avg_Temp=('Temperature', np.mean),
    Max_Temp=('Temperature', np.max),
    Count=('Temperature', 'count'),
    Std_Temp=('Temperature', np.std)
)

print(grouped)
```

```python
df.to_csv("cleaned_weather.csv", index=False)
grouped.to_csv("weather_summary.csv")

print("Files saved successfully")
```
