```python
products = pd.DataFrame({
    'Product': ['Pen', 'Pencil', 'Eraser'],
    'Price': [10, 5, 3]
})

sales = pd.DataFrame({
    'Product': ['Pen', 'Pen', 'Pencil', 'Eraser'],
    'Quantity': [10, 20, 30, 40]
})
```

```python
weather = pd.DataFrame({
    'Date': ['2026-01-01','2026-01-01','2026-01-02','2026-01-02'],
    'City': ['Delhi','Mumbai','Delhi','Mumbai'],
    'Temperature': [28, 30, 27, 31]
})
```

```python
pivot_weather = weather.pivot(
    index='Date',
    columns='City',
    values='Temperature'
)
print(pivot_weather)
```

```python
pivot_table = pd.pivot_table(
    weather,
    values='Temperature',
    index='City',
    aggfunc=[np.mean, np.max, len],
    fill_value=0
)
print(pivot_table)
```

```python
melted = pd.melt(
    pivot_weather.reset_index(),
    id_vars='Date',
    value_vars=['Delhi','Mumbai'],
    var_name='City',
    value_name='Temperature'
)
print(melted)
```

```python
stacked = pivot_weather.stack()
print(stacked)

unstacked = stacked.unstack()
print(unstacked)
```

```python
weather_type = pd.DataFrame({
    'City': ['Delhi','Delhi','Mumbai','Mumbai','Delhi'],
    'Condition': ['Hot','Cold','Hot','Hot','Cold']
})

ct = pd.crosstab(weather_type['City'], weather_type['Condition'])
print(ct)
```

```python
ct_percent = ct.div(np.sum(ct, axis=1), axis=0) * 100
print(ct_percent)
```

```python
jan_sales = pd.DataFrame({
    'Product': ['Pen','Pencil'],
    'Sales': [200, 300],
    'Month': 'Jan'
})

feb_sales = pd.DataFrame({
    'Product': ['Pen','Pencil'],
    'Sales': [250, 280],
    'Month': 'Feb'
})
```

```python
all_sales = pd.concat([jan_sales, feb_sales], ignore_index=True)
print(all_sales)
```

```python
price_list = pd.DataFrame({
    'Product': ['Pen','Pencil'],
    'Price': [10,5]
})

merged_sales = pd.merge(all_sales, price_list, on='Product')
merged_sales['Revenue'] = merged_sales['Sales'] * merged_sales['Price']
print(merged_sales)
```

```python
revenue_pivot = pd.pivot_table(
    merged_sales,
    values='Revenue',
    index='Month',
    columns='Product',
    aggfunc=np.sum,
    fill_value=0
)
print(revenue_pivot)
```

```python
payment = pd.DataFrame({
    'PaymentType': ['Cash','UPI','Cash','Card','UPI'],
    'Segment': ['Retail','Retail','Wholesale','Retail','Wholesale']
})

pay_ct = pd.crosstab(payment['PaymentType'], payment['Segment'])
print(pay_ct)
```

```python
marks = pd.DataFrame({
    'Student':['Amit','Ravi'],
    'Maths':[85,78],
    'Science':[88,82],
    'English':[80,75]
})

long_marks = pd.melt(
    marks,
    id_vars='Student',
    var_name='Subject',
    value_name='Marks'
)
print(long_marks)
```

```python
ct_marks = pd.crosstab(long_marks['Student'], long_marks['Subject'])
ct_percent = ct_marks / np.sum(ct_marks, axis=1).values.reshape(-1,1) * 100
print(ct_percent)
```
