```python
import pandas as pd

df = pd.read_csv('AirQualityUCI.csv', sep=';', decimal=',')
df.head()
```

```python
print(df.columns)
print(df.info())
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df.replace(-200, pd.NA, inplace=True)
```

```python
print(df.columns)
```

```python
df['DateTime'] = pd.to_datetime(
    df['Date'] + ' ' + df['Time'],
    format='%d/%m/%Y %H.%M.%S',
    errors='coerce'
)
```

```python
# Drop original columns
df.drop(['Date', 'Time'], axis=1, inplace=True)
```

```python
df = df.sort_values('DateTime')
```

```python
df = df.fillna(method='ffill')
```

```python
print(df.head())
print(df.isnull().sum())
```

```python
import plotly.express as px

fig = px.line(
    df,
    x='DateTime',
    y=['CO(GT)', 'NOx(GT)', 'T'],
    title='Air Quality Trends Over Time'
)

fig.show()
```

```python
df_daily = df.set_index('DateTime').resample('D').agg({
    'CO(GT)': ['first', 'max', 'min', 'last']
})

df_daily.columns = ['Open', 'High', 'Low', 'Close']
df_daily = df_daily.dropna().reset_index()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(
    x=df_daily['DateTime'],
    open=df_daily['Open'],
    high=df_daily['High'],
    low=df_daily['Low'],
    close=df_daily['Close']
)])

fig.update_layout(title='Daily CO Variation (Candlestick)')
fig.show()
```

```python
import plotly.express as px

df_geo = px.data.gapminder().query("year == 2007")

fig = px.choropleth(
    df_geo,
    locations="iso_alpha",
    color="pop",
    hover_name="country",
    color_continuous_scale="Viridis",
    title="World Population Choropleth Map"
)

fig.show()
```

```python
df_map = px.data.carshare()

fig = px.scatter_map(
    df_map,
    lat="centroid_lat",
    lon="centroid_lon",
    size="car_hours",
    color="peak_hour",
    title="Dot Density Map (Car Usage)"
)

fig.show()
```

```python
fig = px.density_map(
    df_map,
    lat="centroid_lat",
    lon="centroid_lon",
    z="car_hours",
    radius=10,
    title="Hexbin Density Map"
)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'CO Trend Over Time',
        'Daily CO Candlestick',
        'NOx Trend',
        'Temperature Trend'
    )
)
```

```python
# CO Trend
fig.add_trace(
    go.Scatter(
        x=df['DateTime'],
        y=df['CO(GT)'],
        name='CO',
        mode='lines'
    ),
    row=1, col=1
)

# NOx Trend
fig.add_trace(
    go.Scatter(
        x=df['DateTime'],
        y=df['NOx(GT)'],
        name='NOx',
        mode='lines'
    ),
    row=2, col=1
)

# Temperature Trend
fig.add_trace(
    go.Scatter(
        x=df['DateTime'],
        y=df['T'],
        name='Temperature',
        mode='lines'
    ),
    row=2, col=2
)
```

```python
fig.add_trace(
    go.Candlestick(
        x=df_daily['DateTime'],
        open=df_daily['Open'],
        high=df_daily['High'],
        low=df_daily['Low'],
        close=df_daily['Close'],
        name='CO Daily'
    ),
    row=1, col=2
)
```

```python
fig.update_layout(
    height=800,
    title='Air Quality Interactive Dashboard',
    
    # Hover interaction
    hovermode='x unified',
    
    # Enable zoom & pan
    xaxis_rangeslider_visible=False,
    
    # Theme
    template='plotly_dark'
)
```

```python
fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(label="Show All",
                     method="update",
                     args=[{"visible": [True, True, True, True]}]),
                
                dict(label="Only CO",
                     method="update",
                     args=[{"visible": [True, False, False, False]}]),
                
                dict(label="Only NOx",
                     method="update",
                     args=[{"visible": [False, False, True, False]}]),
            ],
            direction="down"
        )
    ]
)
```

```python
fig.show()
```
