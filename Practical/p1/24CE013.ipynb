```python
import numpy as np
import pandas as pd

print("NumPy version:", np.__version__)
print("Pandas version:", pd.__version__)
```

```python
# Integer array
arr_int = np.array([10, 20, 30, 40, 50])

# Float array
arr_float = np.array([2.5, 4.8, 6.1, 8.0])

# Mixed type array
arr_mixed = np.array([10, 3.5, 20, 7.2])

print(arr_int)
print(arr_float)
print(arr_mixed)
```

```python
print("Addition:", arr_int + 5)
print("Subtraction:", arr_int - 5)
print("Multiplication:", arr_int * 2)
print("Division:", arr_int / 2)

print("Mean:", np.mean(arr_int))
print("Max:", np.max(arr_int))
print("Min:", np.min(arr_int))
print("Std Dev:", np.std(arr_int))
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales = pd.Series(arr_int[:5], index=days)

print(sales)
```

```python
# Position based
print(sales[0])
print(sales[1:4])

# Label based
print(sales['Mon'])
print(sales['Mon':'Thu'])
```

```python
print(sales.head())
print(sales.tail())
print(sales.describe())
print(sales.value_counts())
```

```python
students = pd.read_csv("indata.csv")
print(students.head())
```

```python
marks = students[['Maths', 'Science', 'English']]

print("Mean:\n", np.mean(marks))
print("Max:\n", np.max(marks))
print("Min:\n", np.min(marks))
print("Std Dev:\n", np.std(marks))
```

```python
print(marks.describe())
```

```python
students['Percentage'] = (np.sum(marks, axis=1) / 300) * 100
print(students.head())
```

```python
numeric_df = students.select_dtypes(include=np.number)

# Column-wise mean & median
print("Mean:\n", np.mean(numeric_df))
print("Median:\n", np.median(numeric_df))
```

```python
students['Total'] = np.sum(marks, axis=1)
highest_row = students.loc[students['Total'].idxmax()]
print(highest_row)
```

```python
students.to_csv("final_students.csv", index=False)

# Verify
final_df = pd.read_csv("final_students.csv")
print("Rows, Columns:", final_df.shape)
```
