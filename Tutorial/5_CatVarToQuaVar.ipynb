```python
import pandas as pd
```

```python
# CSV load 
df = pd.read_csv("Car_Fuel.csv")
print("Original Data:")
print(df)
```

```python

df = df.join(pd.get_dummies(df['Fuel']))
print("\nAfter get_dummies:")
print(df)
# 1.df['Fuel']
# Selects the Fuel column from the DataFrame df
# 2. pd.get_dummies(df['Fuel'])
# Converts the categorical values into one-hot encoded columns
# 3. df.join(...)
# Joins these new dummy columns to the original DataFrame
# Join happens row-wise using the index
# 4. df = ...
# Stores the updated DataFrame back into df
```

```python
# Alternative - concat() Use:
```

```python

df = pd.read_csv("Car_Fuel.csv")

# Create Dummy variables 
dummy_vars = pd.get_dummies(df['Fuel'], prefix='Fuel')

# Concat with Original df 
df = pd.concat([df, dummy_vars], axis=1)

print(df)


# df['Fuel']
# Selects the Fuel column from the DataFrame.
# This column contains categorical data (e.g., Petrol, Diesel, CNG).

# pd.get_dummies(...)
# Converts each unique category into a separate binary column (one-hot encoding).

# prefix='Fuel'
# Adds Fuel_ before each new column name to avoid confusion or name clashes.

#---------------------->>>

# pd.concat([...])
# Combines multiple DataFrames.

# [df, dummy_vars]
# Takes the original DataFrame and the dummy variables DataFrame.

# axis=1
# Joins them column-wise (side by side).
```
