```python
import pandas as pd
import numpy as np

# Create dataset
np.random.seed(42)

data = {
    "Student_ID": range(1, 101),
    "Marks": np.random.randint(40, 100, 100),
    "Department": np.random.choice(["CS", "IT", "EC"], 100)
}

df = pd.DataFrame(data)
df.head()
```

```python
# Simple random sampling (select 10 students) -> work on indexing value
simple_random_sample = df.sample(n=10, random_state=1)
simple_random_sample

# sample(n=10) → randomly selects 10 rows
# No bias
# Best for small, uniform population
```

```python
# Systematic sampling ->work on indexing value
k = 13  # every 10th student
systematic_sample = df.iloc[::k]
systematic_sample
len(systematic_sample)
# print(systematic_sample)

# iloc[::10] → jump of 10
# Very fast
# Used in ordered data
```

```python
# Stratified sampling (equal samples from each department) we are targating column , work on catagorical value
stratified_sample = df.groupby("Department", group_keys=False).apply(
    lambda x: x.sample(5, random_state=1)
)

stratified_sample

# Population divided into homogeneous groups (strata)
# groupby("Department") → create strata
# sample(5) → sample from each group
# Ensures representation
```

```python
# Stratified random sampling (proportional)
stratified_random_sample = df.groupby("Department", group_keys=False).apply(
    lambda x: x.sample(frac=0.1, random_state=42)
)

stratified_random_sample
```

```python
# Select one cluster randomly
selected_cluster = np.random.choice(df["Department"].unique())
cluster_sample = df[df["Department"] == selected_cluster]

selected_cluster, cluster_sample


# Divide population into clusters
# Select entire cluster(s)
# One department selected
# All students from that department included
# Cost-effective, but less accurate
```

```python
# All available clusters
clusters = df["Department"].unique()
clusters
```

```python
# Select 2 clusters randomly
selected_clusters = np.random.choice(clusters, size=3, replace=False)
selected_clusters
```

```python
# Multiple cluster sampling
multi_cluster_sample = df[df["Department"].isin(selected_clusters)]
multi_cluster_sample
```
