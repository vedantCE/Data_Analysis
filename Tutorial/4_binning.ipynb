```python
# Binning શું છે? 
# Binning એટલે continuous data ને categories/groups માં વહેંચવાની process. આને "વર્ગીકરણ" અથવા "જૂથબંધી" કહી શકાય.

# જેમ તમે પેન્સિલને પેન્સિલ બોક્સમાં અને કપડાંને અલગ અલગ drawers માં ગોઠવો છો, તેમ binning data ને અલગ અલગ categories માં ગોઠવે છે

# Binning (also called discretization) is a data preprocessing technique that converts continuous numerical data into discrete categories or bins. This helps simplify data analysis, reduce noise, and better understand data distribution patterns, making it particularly useful when working with features like product prices, ages, or income levels
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #just for understand no required
```

```python
df=pd.read_csv("product_prices.csv") # reading of csv , header avalaiable che to aapde header=None no lakhyu
```

```python
# Sample price data
df['Price_Category'] = pd.cut(df['Price'], 
                               bins=5, 
                               labels=['Budget', 'Mid-Range', 'Premium', 'Luxury',"Extra Premium"])

# bins=4 - This tells pandas to create 4 equal-width bins (intervals). Pandas automatically calculates the range (max - min) and divides it into 4 equal parts. For example, if prices range from ₹500 to ₹50,000, each bin width would be approximately ₹12,375.
```

```python
print(df[['Product_Name', 'Price', 'Price_Category']])
```

```python
# Step 4: Graph માં જુઓ khali samajava matej che ke aapde data ne keltli sarl tathi joi shakia  
plt.figure(figsize=(10, 6))
df['Price_Category'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Product Distribution by Price Category', fontsize=16, fontweight='bold')
plt.xlabel('Price Category', fontsize=12)
plt.ylabel('Number of Products', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Equal-Frequency Binning (pd.qcut)
# Equal-frequency binning creates bins with approximately equal numbers of observations using quantiles. The pd.qcut() function ensures each bin contains roughly the same number of data points, making it ideal for evenly distributed analysis.
```

```python
# Create 4 bins with equal number of observations
df['Price_Category_EqualFreq'] = pd.qcut(df['Price'], 
                                          q=4, 
                                          labels=['Q1-Budget', 'Q2-Economy', 
                                                  'Q3-Standard', 'Q4-Premium'])
```

```python
print(df[['Product_Name', 'Price', 'Price_Category_EqualFreq']])
```
